```python
# Task-1 (a) #
def handleFiles(inputFile, outputFile):
  with open(inputFile, 'r') as infile:

    T = int(infile.readline())
    res = []

    for _ in range(T):
      n = int(infile.readline())
      if n % 2 == 1:
        res.append(f"{n} is an Odd number.")
      elif n % 2 == 0:
        res.append(f"{n} is an Even number.")

  with open(outputFile, 'w') as outfile:
    for i in res:
      outfile.write(i + "\n")

inputFile = "/content/drive/MyDrive/Colab Notebooks/Lab 221/Lab 1/input1a.txt"
outputFile = "/content/drive/MyDrive/Colab Notebooks/Lab 221/Lab 1/output1a.txt"
handleFiles(inputFile, outputFile)
```

```python
# Task - 2
def handleFiles(inputFile, outputFile):
    with open(inputFile, 'r') as infile:
        t = int(infile.readline())
        lst = list(map(int, infile.readline().split()))
        res = []

        def bubbleSort(arr):
            for i in range(len(arr) - 1):
                check = False
                for j in range(len(arr) - i - 1):
                    if arr[j] > arr[j + 1]:
                        arr[j], arr[j + 1] = arr[j + 1], arr[j]
                        check = True  # if it triggers, that means one swap is just performed.
                if check == False:  # no swapping done in the second for loop means the array is already sorted, so we don't need to check further. now we can return from the function.
                    return arr
            return arr

        res = bubbleSort(lst)

    with open(outputFile, 'w') as outfile:
        for i in res:
            outfile.write(str(i) + " ")


inputFile = "/content/drive/MyDrive/Colab Notebooks/Lab 221/Lab 1/input1a.txt"
outputFile = "/content/drive/MyDrive/Colab Notebooks/Lab 221/Lab 1/output1a.txt"
handleFiles(inputFile, outputFile)


# My observation and explanation:
# Accoring to the question the best case will be if i give a sorted array then it should return after O(N) running time.

# My Approach:
# In bubble sort its compare two element and then swap the elements if first element is greater than the second element. And its happen in the inside for loop.
# For this code *(for j in range(len(arr) - i - 1): )* inside this loop the swapping is happening. So outside the loop i take a varibale "check" and initialize this as False.
# Inside the second for loop, where this code is performing swapping i wrote * check = True *. That's mean if a single swap is done the previous initialze check will set as True.
# After one interation through the entire loop i'm checking if the check variable is still False or not. If its still false it will return the array from the fuction because our given array is already sorted.

# Time Complexity Analysis:
# If the array is sorted alreay for the inside for loop it will check and run N times. Then it will check the break condition and return the array from the funciton.
# So for the best case the time complexity will be O(N)

```

```python
# task - 3
def handleFiles(inputFile, outputFile):
    with open(inputFile, 'r') as infile:

        # Selection sort function for sorting by scores in descending order.
        def selection_sort(A):
            for i in range(len(A) - 1):
                max_idx = i
                for j in range(i + 1, len(A)):
                    if A[max_idx][0] < A[j][0]:
                        max_idx = j
                A[i], A[max_idx] = A[max_idx], A[i]

        # Selection sort function for sorting by IDs in ascending order within segments.
        def id_selection_sort(A, start, end):
            for i in range(start, end):
                min_idx = i
                for j in range(i + 1, end + 1):
                    if A[min_idx][1] > A[j][1]:
                        min_idx = j
                A[i], A[min_idx] = A[min_idx], A[i]

        N = int(infile.readline())
        id = list(map(int, infile.readline().split()))
        num = list(map(int, infile.readline().split()))
        res = []
        stu = []
        for i in range(N):
          stu.append((num[i], id[i]))

        selection_sort(stu)

        # same element paile oi particular segment list ta ascending order e sort dibo
        i = 0
        while i < len(stu):
            start = i
            while i < len(stu) - 1 and stu[i][0] == stu[i + 1][0]:
                i += 1
            end = i
            id_selection_sort(stu, start, end)
            i += 1

        # print(stu)
        for i in stu:
          res.append(f"ID: {i[1]} Mark: {i[0]}")



    with open(outputFile, 'w') as outfile:
        for id_str in res:
            outfile.write(id_str + "\n")

# Example usage
inputFile = "/content/drive/MyDrive/Colab Notebooks/Lab 221/Lab 1/input1a.txt"
outputFile = "/content/drive/MyDrive/Colab Notebooks/Lab 221/Lab 1/output1a.txt"
handleFiles(inputFile, outputFile)


# Time Complexity analysis:
# The first sorting step by scores numbers is O(N^2)
# The second sorting step by IDs within segments is another O(N^2) in the worst case.
# Therefore, the combined time complexity is O(N ^ 2) + O(N ^ 2) = O(N ^ 2)
```

```python
# Task - 4
def handleFiles(inputFile, outputFile):
    with open(inputFile, 'r') as infile:

        # all functions
        def time_to_minutes(time):
            hours, minutes = map(int, time.split(':'))
            return hours * 60 + minutes

        N = int(infile.readline())
        lst = []
        for i in range(N):
            temp = infile.readline()
            lst.append(temp)

        # Selection sort
        for i in range(len(lst)):
            min_idx = i
            for j in range(i + 1, len(lst)):
                if lst[j] < lst[min_idx]:
                    min_idx = j
            lst[i], lst[min_idx] = lst[min_idx], lst[i]

        res = []
        new_lst = []
        for i in lst:
            new_lst.append(i.split(" "))

        for i in range(len(new_lst)):
            for j in range(i + 1, len(new_lst)):
                if new_lst[i][0] == new_lst[j][0]:
                    if time_to_minutes(new_lst[j][6]) > time_to_minutes(new_lst[i][6]):
                        new_lst[i], new_lst[j] = new_lst[j], new_lst[i]

        for i in new_lst:
            temp = ""
            for j in i:
                temp += str(j) + " "
            res.append(temp)
            temp = ""
        print(res)

    with open(outputFile, 'w') as outfile:
        for id_str in res:
            outfile.write(id_str)

# Example usage
inputFile = "/content/drive/MyDrive/Colab Notebooks/Lab 221/Lab 1/input1a.txt"
outputFile = "/content/drive/MyDrive/Colab Notebooks/Lab 221/Lab 1/output1a.txt"
handleFiles(inputFile, outputFile)

# Time Complexity analysis:
# Selection sort: O(N^2)
# Nested loop for custom sorting within new_lst: O(N^2)
# The dominant term in the overall time complexity is O(N^2) due to the selection sort implementation and the nested loop for sorting described in the task.
# Overall Time Complexity: O(N^2)
```
